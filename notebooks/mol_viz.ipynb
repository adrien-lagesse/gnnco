```python
import gnnco
```

```python
graph, coordinates = gnnco.chem.smiles_to_graph("C[C@H](CCCC(C)C)[C@H]1CC[C@@H]\\2[C@@]1(CCC/C2=C\\C=C/3\\C[C@H](CCC3=C)O)C ")
```

```python
gnnco.visualization.plot_graph(
    graph,
    coordinates=coordinates,
)
```

```python
corrupted_graph = gnnco.random.bernoulli_corruption(graph.to_batch(), 0.1, type='node_normalized')[0]
```

```python
gnnco.visualization.compare_graphs(graph, corrupted_graph, coordinates=coordinates)
```
